#### 1. Which establishments have a hygiene score equal to 20?

#### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_2 = {"LocalAuthorityName": "City of London Corporation", "RatingValue": {"$gte": 4}}

# Use count_documents to display the number of documents in the result
print('Number of establishements with a hygiene score of 4 or greater in London: ', establishments.count_documents(query_2))

# Display the first document in the results using pprint
pprint(establishments.find_one(query_2))
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(establishments.find(query_2))
                         
# Display the number of rows in the DataFrame
print('Number of rows in the DataFrame: ', len(london_df))

# Display the first 10 rows of the DataFrame
print(london_df.head(10))
```

#### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

# Search for enteries with a rating of 5 and latitude and longitude within 0.01 degree
query_3 = {"RatingValue": 5, "Latitude": 
           {"$gte": latitude - degree_search, "$lte": latitude + degree_search}, 
           "Longitude": 
           {"$gte": longitude - degree_search, "$lte": longitude + degree_search}}
sort = [("RatingValue", 1)]
limit = 5

# Print the results
print('Number of establishements with a hygiene score of 5 within 0.01 degree of the latitude and longitude: ', establishments.count_documents(query_3))
```

```python
# Convert result to Pandas DataFrame
degree_df = pd.DataFrame(list(establishments.find(query_3).sort(sort).limit(limit)))
```

#### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match = {"$match": {"RatingValue": 0}}
group = {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}}
sort = {"$sort": {"count": -1}}

# Aggregate the results
pipeline = [match, group, sort]
result = establishments.aggregate(pipeline)

# Print the number of documents in the result
print('Number of establishements with a hygiene score of 0: ', establishments.count_documents({"RatingValue": 0}))

# Print the first 10 results
pprint(list(result)[:10])
```

```python
# Convert the result to a Pandas DataFrame
zero_df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
print('Number of rows in the DataFrame: ', len(zero_df))

# Display the first 10 rows of the DataFrame
pprint(zero_df.head(10))
```
